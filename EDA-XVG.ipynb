```python
twts.favorite = twts.favorite.apply(lambda x: int(x) if len(x) != 0)
twts.reply = twts.reply.apply(lambda x: int(x) if x != '')
twts.retweet = twts.retweet.apply(lambda x: int(x) if x != '')
```

```python
def convert(x):
    if x != '' and x.endswith('K') == False:
            return int(x)
    else:
        if x.endswith('K'):
            return float(x.replace('K',''))*1000
        else:
            return 0
```

```python
twts.favorite = twts.favorite.apply(convert)
twts.reply = twts.reply.apply(convert)
twts.retweet = twts.retweet.apply(convert)
```

```python
twts['exposure'] = twts.favorite + twts.reply.apply(lambda x:x*10) + twts.retweet.apply(lambda x:x*5)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
class collocation(BaseEstimator, TransformerMixin):
    def __init__(self,common_terms,all_token):
        self.common_terms = common_terms
        self.all_token = all_token
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        collocationed_sentences = []
        phrases = Phrases(self.common_terms, self.common_terms)
        colloc = Phraser(phrases)
        for doc in nlp.pipe(X, n_threads=8):
            collocationed_sentences.append(' '.join(colloc([t.lower_ for t in doc if not t.is_punct])))
        return collocationed_sentences
```

```python
stop_words = {'computer','you','for','a', 'sun','and', 'user','to','it','these','i', 'week','flash','flashlight','light','day','deliver','ups','post'}
class reviewtextFilter(BaseEstimator, TransformerMixin):
    def __init__(self, pos_to_keep=('NOUN', 'PROPN', 'ADJ', 'VERB'), stop_words=None):
        self.pos_to_keep = pos_to_keep
        self.stop_words = stop_words
        
    def fit(self, X, y=None):
        return self
        
    def transform(self, X):
        preprocessed_sentences = []
        
        for doc in nlp.pipe(X, n_threads=8):
            if self.stop_words is not None:
                keep_tokens_string = ' '.join([t.lemma_ for t in doc if t.pos_ in self.pos_to_keep and t.lemma_ not in self.stop_words])
            else:
                keep_tokens_string = ' '.join([t.lemma_ for t in doc if t.pos_ in self.pos_to_keep])
            preprocessed_sentences.append(keep_tokens_string)
    
        return preprocessed_sentences
```
